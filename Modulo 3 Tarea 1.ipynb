```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DelayedFlights.csv', index_col=0)
df.head()
```

```python
values_UC = df['UniqueCarrier'].value_counts()
```

### Graficar Variable Categorica


```python
plt.figure(figsize=(10,10))
y = values_UC
mylabels = y.index
plt.title('UniqueCarrier', fontsize=20)
explode = [0]* len(mylabels)
explode[0] = 0.1
explode[-1]= 0.1
plt.pie(y, labels= mylabels, explode=explode, shadow=True, autopct='%.2f%%', startangle=0)

plt.legend(bbox_to_anchor=(1,1),loc=2,fontsize=12, title="Aerolineas:")
plt.savefig('pieplot_UniqueCarrier.png')
plt.show()
```

### Una variable numerica

```python
### Vamos a comprobar la distribucion de la data para poder tener una mejor perspectiva.

fig = plt.figure(figsize=(8,8))

sns.boxplot(x=df.ArrDelay, y=df.UniqueCarrier)
plt.title('Box Plot: Total flights in Delays', fontsize=14)
plt.xlabel("Arrival delay in minutes",fontsize=12)


plt.savefig('boxPlot_ArrDelay.png')

plt.show()
```

```python
data = df['ArrDelay']


fig,axes = plt.subplots(1, 2, figsize=(15,8))

sns.distplot(data, hist=False, ax = axes[0])
axes[0].set(xlabel ='Delay in minutes', ylabel ='Density',
       title ='Arrival Delay')\


x = df['ArrDelay']
plt.hist(x, bins=40)

plt.title("Arrival Delay",fontsize=20)
plt.xlabel("Delay in minutes", fontsize=14)
plt.ylabel("Numbers of flights", fontsize=14)
plt.xlim(None, df['ArrDelay'].max())
plt.grid(color='g', linestyle='-', linewidth=0.5)

plt.savefig("subplot_arrdelay.png")



plt.show()
```

### 3 Variables Numericas


```python
data = df[["UniqueCarrier","ArrDelay","DepDelay"]]
data = data.groupby("UniqueCarrier")[["ArrDelay","DepDelay"]].sum()
data= data.sort_values(by="ArrDelay", ascending=False)
labels = data.index
```

```python
fig, ax = plt.subplots(figsize=(10,8))

ax.bar(data.index, data["ArrDelay"],label="Arrival delay")
ax.bar(data.index, data["DepDelay"], bottom=data["ArrDelay"],label= "Departure delay",color='lime')
ax.set_title("Total Delay by Airline",fontsize=20)
ax.set_ylabel("Total delay in minutes", fontsize=14)
ax.set_xlabel("Airlines code", fontsize=14)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.yaxis.grid(True)
ax.xaxis.grid(False)

plt.legend(loc="best", title="Variables:")

plt.savefig('stacker_bar.png')

plt.show()
```

```python
data = df[["UniqueCarrier","ArrDelay","DepDelay","AirTime"]]
data = data.groupby("UniqueCarrier")[["ArrDelay","DepDelay","AirTime"]].sum()

plt.figure(figsize = (15,10))
ax = sns.scatterplot(data=data, x="ArrDelay", y="DepDelay",\
                    size="AirTime", sizes=(500, 1500),\
                    alpha = 0.5,
                    hue="UniqueCarrier",
                    palette="rainbow",
                    edgecolor="black",
                    )



for line in range(0,data.shape[0]):
     ax.text(data.ArrDelay[line], data.DepDelay[line], data.index[line], \
        horizontalalignment='center', size='small')

plt.legend(title="Size")    
plt.legend(bbox_to_anchor=(1, 1),loc='best', fontsize=11, borderaxespad=2)

plt.title("Arrival & Departure Delay and Air Time by Airline ", fontsize=20)
plt.xlabel("Sum of Arrival delay in minutes", fontsize= 14, color="r")
plt.ylabel("Sum Departure delay in minutes", fontsize= 14, color="r")


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.yaxis.grid(True)
ax.xaxis.grid(True)
plt.savefig('bubble_chart.png')
plt.show()
```
